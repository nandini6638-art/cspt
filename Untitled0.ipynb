```python
def add(a, b):
    print("Result =", a + b)

def subtract(a, b):
    print("Result =", a - b)

def multiply(a, b):
    print("Result =", a * b)

def divide(a, b):
    if b == 0:
        print("Cannot divide by zero")
    else:
        print("Result =", a / b)

a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

choice = input("Enter your choice (1-4): ")

if choice == "1":
    add(a, b)
elif choice == "2":
    subtract(a, b)
elif choice == "3":
    multiply(a, b)
elif choice == "4":
    divide(a, b)
else:
    print("Invalid choice")
```
